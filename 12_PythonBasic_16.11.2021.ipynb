```python
answer = 22

print("Lets's play a game. Guess the number")

while True:
    guess = int(input("enter your guess : "))

    if guess > answer :
        print("little lower")
       
    elif guess < answer :
        print("little higher")
        
    else:
        print("congratz")
        break
```

```python
sentence = input("give me a sentence")
word_list = sentence.split()
i = 0
longest = 0
print(word_list)
while i < len(word_list):
    if len(word_list[i]) > longest :
        longest = len(word_list[i])
    
    i +=1

print("the length of the longest word : ", longest)

#len(max(word_list, key = len))
```
