```python
def fakto(var):
    result = 1
    for i in range(1,var +1):
        result *= i
    return result
```

```python
fakto(0)
```

```python
fakto(1)
```

```python
fakto(5)
```

```python
def fakto2(var) :
    if var == 1 or var == 0 :
        return 1
    else:
        result = var * fakto2(var-1)
        return result
```

```python
fakto2(3000)
```

```python
fakto2(2000)
```

```python
fakto(2000)
```
